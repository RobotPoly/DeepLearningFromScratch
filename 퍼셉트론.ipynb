#### 퍼셉트론이란?
0과 1의 신호의 흐름이다.

그림 처럼 X값을 받아 W(가충치)를 받은 뒤에 Y값을 반환한다.

여기서 신호가 즉 y로 가는 값이 특정 값(임계값)을 넘으면 1을 받고 넘지 못하면 0을 받는다고 하자. 
그렇다면 식으로 표현했을 때 다음과 같다.


-------------------------

#### AND, NAND, OR, XOR에 대해 알아보자

###### AND 게이트 
- AND 게이트는 아래와 같이 출력된다.
- X값이 모두 1이어야 1이라는 y값을 받을 수 있다.


```python
#and게이트 코드
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7 #x값이 모두 1이어야 y값이 1인 값을 반환하도록 설정
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

###### NAND 게이트
- AND게이트를 반대로 생각하면 된다.
- X값이 모두 1일 때만 0을 반환하고 나머지 경우는 모두 1이다.


```python
import numpy as np
#NAND게이트 코드
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

###### OR 게이트
- X값이 모두 0일때만 0을 반환하고 나머지는 1이다.


```python
#OR게이트 코드
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

###### XOR게이트
- 식으로 표현이 불가능하다.
- 하지만 AND OR NAND를 합치면 구현이 가능하다.
- 비선형 형태이다.



- AND OR NAND 를 사용해서 XOR을 구현해 보면 아래와 같다.
- 여기서 두 경로를 통해 값을 얻었음으로 다중 퍼셉트론 이라고 할 수 있다.



```python
#XOR 코드 구현
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

XOR을 통해 알 수 있는 것은 우리가 단층(OR, AND, NAND)을 통해 얻을 수 없는 구현을 다층을 통해 얻는 모습을 볼 수 있다.
